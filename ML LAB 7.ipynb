```
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # visualization
import matplotlib.pyplot as plt

%matplotlib inline
```

```
# load database
data = pd.read_csv("/content/drive/My Drive/College/6th Sem/ML/Lab/nba_logreg.csv")
data.dropna(inplace=True)
data.sample(n=5)
```

```
data = data.drop(columns=["name"])
df_norm = data[["gp", "min", "pts", "fgm", "fga", "fg","3p_made", "3pa", "3p", "ftm", "fta", "ft", "oreb", "dreb", "reb", "ast", "stl", "blk", "tov"]].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
df_norm.head()
```

```
target = data[['target_5yrs']]
target.sample(n=5)
```

```
df = pd.concat([df_norm, target], axis=1)
df.sample(n=5)
```

```
train_test_per = 80/100.0
df['train'] = np.random.rand(len(df)) < train_test_per
df.sample(n=5)
```

```
train = df[df.train == 1]
train = train.drop('train', axis=1).sample(frac=1)
train.sample(n=5)
```

```
test = df[df.train == 0]
test = test.drop('train', axis=1)
test.sample(n=5)
```

```
X = train.values[:,:19]
X[:5]
```

```
targets = [[1,0],[0,1]]
y = np.array([targets[int(x)] for x in train.values[:,19:20]])
y[:5]
```

```
num_inputs = len(X[0])
hidden_layer_neurons1 = 6
np.random.seed(4)
w1 = 2*np.random.random((num_inputs, hidden_layer_neurons1)) - 1
print(w1)
num_outputs = len(y[0])
w2 = 2*np.random.random((hidden_layer_neurons1, num_outputs)) - 1
w2
```

```
learning_rate = 0.2 # slowly update the network
for epoch in range(500000):
    l1 = 1/(1 + np.exp(-(np.dot(X, w1)))) # sigmoid function
    l2 = 1/(1 + np.exp(-(np.dot(l1, w2))))
    er = (abs(y - l2)).mean()
    l2_delta = (y - l2)*(l2 * (1-l2))
    l1_delta = l2_delta.dot(w2.T) * (l1 * (1-l1))
    w2 += l1.T.dot(l2_delta) * learning_rate
    # print('LLLL :', w2)
    w1 += X.T.dot(l1_delta) * learning_rate
print('Error:', er)
```

```
X = test.values[:,:19]
y = np.array([targets[int(x)] for x in test.values[:,19:20]])

l1 = 1/(1 + np.exp(-(np.dot(X, w1))))
l2 = 1/(1 + np.exp(-(np.dot(l1, w2))))
np.round(l2,3)
```

```
yp = np.argmax(l2, axis=1) # prediction
res = yp == np.argmax(y, axis=1)
correct = np.sum(res)/len(res)

testres = test[['target_5yrs']]

testres['Prediction'] = yp
# testres['Prediction'] = testres['Prediction'].replace([0.0,1.0], ['No','Yes'])

print(testres)
print('Correct:',sum(res),'/',len(res), ':', (correct*100),'%')
```
